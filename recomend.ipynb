Recomendation Function and importing data plus index

```python
import pandas as pd

import faiss
import requests
import numpy as np

from my_util_func import get_netflix_dataframe, get_embedding_from_representation

index = faiss.read_index('index')
df = get_netflix_dataframe()

def recomend(textual_representation, top_n):
    # embed info
    embedding = get_embedding_from_representation(textual_representation)
    embedding = np.array([embedding], dtype='float32')

    # search
    D, I = index.search(embedding, top_n)

    best_matches = np.array(df['textual_representation'])[I.flatten()]
    for match in best_matches:
        print('Watch Next?')
        print(match)
        print()
```
